```python
import pandas as pd
import numpy as np

# ТОЧНО ЗАГРУЗИ ИСХОДНЫЙ CSV
data = pd.read_csv('turtles.csv', sep='\t', decimal=',')
data.head()
```

```python
# ФИЛЬТР ТОЛЬКО ЧЕРЕПАХ
data['binomial_name'] = data['binomial_name'].str.lower().str.replace(' ', '_')
data = data[data['binomial_name'] == 'chelonia_mydas']
data
```

```python
# УДАЛЕНИЕ ДУБЛЕКАТОВ
data = data.drop_duplicates()
data = data.drop_duplicates(subset=['registration number'], keep='first')
```

```python
# УДАЛЕНИЕ ПРОПУСКОВ В ТАРГЕТЕ
data = data.dropna(subset=['weight'])
```

```python
# ИСПРАВЛЕНИЕ ПАНЦИРЯ — ТОЛЬКО ТАК, ТОЛЬКО ОДИН РАЗ!
data['shell_length'] = data['shell_length'].apply(lambda x: x / 10 if x > 1500 else x)
data['shell_width'] = data['shell_width'].apply(lambda x: x / 10 if x > 1320 else x)
```

```python
# СОЗДАЁМ ПЛОЩАДЬ
data['shell_area'] = data['shell_length'] * data['shell_width']
```

```python
data.drop(['binomial_name', 'registration number'],axis=1, inplace=True)
```

```python
# Удаляем строки, где вес равен 0 (ошибка измерения)
data = data[data['weight'] > 1.0]
def prepare_features(feature):
    if feature == 'all_features':
            return data.copy()
            
    elif 'correlation' in feature:
            per = float(feature.split('>')[1])
            dataset = pd.DataFrame(data.corr()['weight'].drop('weight')).reset_index()
            dataset.columns = ['col', 'cor']
            use_features = dataset[dataset['cor'] >= per]['col'].tolist()
            return data[use_features]
        
    elif 'top' in feature:
            top = int(feature.split('_')[1])
            dataset = pd.DataFrame(data.corr()['weight'].drop('weight')).reset_index().sort_values(by='weight', ascending=False)[:top]
            dataset.columns = ['col', 'cor']
            use_features = dataset['col'].tolist()
            return data[use_features]
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Lasso, Ridge, SGDRegressor
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_score, train_test_split
import time
import warnings
warnings.filterwarnings('ignore')

class LinearModelTrainer:
    """
    Класс для обучения и сравнения различных линейных моделей
    """
    
    def __init__(self, data, target_column='weight'):
        """
        Инициализация класса
        
        Parameters:
        -----------
        data : DataFrame
            Исходные данные
        target_column : str
            Название целевой переменной
        """
        self.data = data.copy()
        self.target_column = target_column
        self.X = None
        self.y = None
        self.models = {}
        self.results = []
        
        # Очищаем данные от некорректных значений
        if target_column in data.columns:
            self.data = self.data[self.data[target_column] > 1.0]
    
    def prepare_features(self, feature_strategy, return_data=False):
        """
        Отбор признаков по заданной стратегии
        
        Parameters:
        -----------
        feature_strategy : str
            Стратегия отбора признаков
        return_data : bool
            Вернуть ли DataFrame вместо X, y
        
        Returns:
        --------
        X, y или DataFrame
        """
        if feature_strategy == 'all_features':
            X_data = self.data.drop(self.target_column, axis=1)
            y = self.data[self.target_column]
            
        elif feature_strategy.startswith('correlation>'):
            threshold = float(feature_strategy.split('>')[1])
            corr_values = self.data.corr()[self.target_column].drop(self.target_column).abs()
            use_features = corr_values[corr_values >= threshold].index.tolist()
            
            if len(use_features) == 0:
                print(f"Предупреждение: нет признаков с корреляцией > {threshold}")
                use_features = corr_values.nlargest(1).index.tolist()
            
            X_data = self.data[use_features]
            y = self.data[self.target_column]
            
        elif feature_strategy.startswith('top_'):
            top_k = int(feature_strategy.split('_')[1])
            corr_values = self.data.corr()[self.target_column].drop(self.target_column).abs()
            top_features = corr_values.nlargest(top_k).index.tolist()
            
            X_data = self.data[top_features]
            y = self.data[self.target_column]
            
        elif feature_strategy == 'no_correlated':
            # Убираем сильно коррелирующие признаки (>0.95)
            X_temp = self.data.drop(self.target_column, axis=1)
            corr_matrix = X_temp.corr().abs()
            upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
            
            to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
            use_features = [col for col in X_temp.columns if col not in to_drop]
            
            X_data = self.data[use_features]
            y = self.data[self.target_column]
            
        else:
            raise ValueError(f"Неизвестная стратегия: {feature_strategy}")
        
        if return_data:
            return pd.concat([X_data, y], axis=1)
        
        self.X = X_data
        self.y = y
        return X_data, y
    
    def scale_features(self, X, method='none'):
        """
        Масштабирование признаков
        
        Parameters:
        -----------
        X : array-like
            Признаки
        method : str
            Метод масштабирования ('none', 'standard', 'minmax', 'robust')
        
        Returns:
        --------
        X_scaled : array-like
        """
        if method == 'none':
            return X
        
        scalers = {
            'standard': StandardScaler(),
            'minmax': MinMaxScaler(),
            'robust': RobustScaler()
        }
        
        if method in scalers:
            return scalers[method].fit_transform(X)
        else:
            raise ValueError(f"Неизвестный метод масштабирования: {method}")
    
    def train_linear_regression(self, X_train, y_train, **kwargs):
        """
        Обучение LinearRegression
        """
        model = LinearRegression(**kwargs)
        model.fit(X_train, y_train)
        return model
    
    def train_lasso(self, X_train, y_train, alpha=1.0, **kwargs):
        """
        Обучение Lasso (L1-регуляризация)
        """
        model = Lasso(alpha=alpha, random_state=42, **kwargs)
        model.fit(X_train, y_train)
        return model
    
    def train_ridge(self, X_train, y_train, alpha=1.0, **kwargs):
        """
        Обучение Ridge (L2-регуляризация)
        """
        model = Ridge(alpha=alpha, random_state=42, **kwargs)
        model.fit(X_train, y_train)
        return model
    
    def train_sgd(self, X_train, y_train, loss='squared_error', penalty='l2', 
                  alpha=0.0001, max_iter=1000, **kwargs):
        """
        Обучение SGDRegressor
        """
        model = SGDRegressor(
            loss=loss,
            penalty=penalty,
            alpha=alpha,
            max_iter=max_iter,
            random_state=42,
            **kwargs
        )
        model.fit(X_train, y_train)
        return model
    
    def evaluate_model(self, model, X_test, y_test, model_name, params):
        """
        Оценка модели на тестовых данных
        """
        y_pred = model.predict(X_test)
        
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        # Количество ненулевых коэффициентов (для Lasso)
        n_nonzero = np.sum(model.coef_ != 0) if hasattr(model, 'coef_') else len(model.coef_)
        
        result = {
            'model': model_name,
            'params': str(params),
            'mse': mse,
            'rmse': rmse,
            'mae': mae,
            'r2': r2,
            'n_features': X_test.shape[1],
            'n_nonzero_coef': n_nonzero
        }
        
        return result
    
    def run_experiment(self, feature_strategies, scaling_methods, 
                       train_test_split_ratio=0.2):
        """
        Запуск полного эксперимента со всеми комбинациями
        """
        # Разделяем данные на train/test один раз
        X_full, y_full = self.prepare_features(feature_strategies[0])
        X_train_full, X_test_full, y_train_full, y_test_full = train_test_split(
            X_full, y_full, test_size=train_test_split_ratio, random_state=42
        )
        
        for feature_strategy in feature_strategies:
            print(f"\n{'='*60}")
            print(f"Отбор признаков: {feature_strategy}")
            print(f"{'='*60}")
            
            # Получаем данные для текущей стратегии
            X, y = self.prepare_features(feature_strategy)
            
            # Разделяем на train/test
            X_train, X_test, y_train, y_test = train_test_split(
                X, y, test_size=train_test_split_ratio, random_state=42
            )
            
            for scaling_method in scaling_methods:
                print(f"\n--- Масштабирование: {scaling_method} ---")
                
                # Масштабируем признаки
                X_train_scaled = self.scale_features(X_train, scaling_method)
                X_test_scaled = self.scale_features(X_test, scaling_method)
                
                # 1. Linear Regression
                print("  Обучение LinearRegression...")
                lr = self.train_linear_regression(X_train_scaled, y_train)
                result = self.evaluate_model(lr, X_test_scaled, y_test, 
                                            'LinearRegression', {'scaling': scaling_method})
                result['feature_strategy'] = feature_strategy
                result['scaling'] = scaling_method
                self.results.append(result)
                
                # 2. Lasso с разными alpha
                for alpha in [0.001, 0.01, 0.1, 1.0, 10.0]:
                    print(f"  Обучение Lasso (alpha={alpha})...")
                    lasso = self.train_lasso(X_train_scaled, y_train, alpha=alpha)
                    result = self.evaluate_model(lasso, X_test_scaled, y_test,
                                                'Lasso', {'alpha': alpha, 'scaling': scaling_method})
                    result['feature_strategy'] = feature_strategy
                    result['scaling'] = scaling_method
                    self.results.append(result)
                
                # 3. Ridge с разными alpha
                for alpha in [0.001, 0.01, 0.1, 1.0, 10.0]:
                    print(f"  Обучение Ridge (alpha={alpha})...")
                    ridge = self.train_ridge(X_train_scaled, y_train, alpha=alpha)
                    result = self.evaluate_model(ridge, X_test_scaled, y_test,
                                                'Ridge', {'alpha': alpha, 'scaling': scaling_method})
                    result['feature_strategy'] = feature_strategy
                    result['scaling'] = scaling_method
                    self.results.append(result)
                
                # 4. SGDRegressor с разными loss функциями
                for loss in ['squared_error', 'huber']:
                    for penalty in ['l1', 'l2', 'elasticnet']:
                        print(f"  Обучение SGD (loss={loss}, penalty={penalty})...")
                        sgd = self.train_sgd(X_train_scaled, y_train, 
                                           loss=loss, penalty=penalty, alpha=0.0001)
                        result = self.evaluate_model(sgd, X_test_scaled, y_test,
                                                    'SGDRegressor', 
                                                    {'loss': loss, 'penalty': penalty, 'scaling': scaling_method})
                        result['feature_strategy'] = feature_strategy
                        result['scaling'] = scaling_method
                        self.results.append(result)
    
    def get_results_table(self):
        """
        Получение таблицы с результатами
        """
        results_df = pd.DataFrame(self.results)
        return results_df.sort_values('r2', ascending=False)
    
    def get_best_model(self, metric='r2'):
        """
        Получение лучшей модели по выбранной метрике
        """
        results_df = self.get_results_table()
        if metric == 'r2':
            best = results_df.iloc[0]
        else:
            best = results_df.nsmallest(1, metric).iloc[0]
        
        print(f"\n{'='*60}")
        print(f"Лучшая модель по метрике {metric.upper()}:")
        print(f"{'='*60}")
        print(f"Модель: {best['model']}")
        print(f"Параметры: {best['params']}")
        print(f"Стратегия отбора признаков: {best['feature_strategy']}")
        print(f"Масштабирование: {best['scaling']}")
        print(f"R²: {best['r2']:.4f}")
        print(f"MSE: {best['mse']:.4f}")
        print(f"RMSE: {best['rmse']:.4f}")
        print(f"MAE: {best['mae']:.4f}")
        
        return best
    
    def compare_models_summary(self):
        """
        Сравнение моделей (средние показатели по всем экспериментам)
        """
        results_df = self.get_results_table()
        
        summary = results_df.groupby('model').agg({
            'r2': ['mean', 'std', 'max'],
            'rmse': ['mean', 'std', 'min'],
            'mae': ['mean', 'std', 'min']
        }).round(4)
        
        print("\n" + "="*80)
        print("СРАВНЕНИЕ МОДЕЛЕЙ (средние показатели по всем экспериментам)")
        print("="*80)
        print(summary)
        
        return summary



    
    # Запуск экспериментов
    feature_strategies = [
        'all_features',
        'correlation>0.3',
        'top_5',
        'no_correlated'
    ]
    
    scaling_methods = ['none', 'standard', 'minmax']
    
    trainer = data.run_experiment(feature_strategies, scaling_methods)
    
    # Получение результатов
    results = trainer.get_results_table()
    print("\n" + "="*80)
    print("ВСЕ РЕЗУЛЬТАТЫ (топ-10)")
    print("="*80)
    print(results.head(10))
    
    # Лучшая модель
    best = trainer.get_best_model('r2')
    
    # Сравнение моделей
    summary = trainer.compare_models_summary()
```
